```python
# Visualising the data

plt.rcParams['figure.figsize']=(20,10)

plt.subplot(2,4,1)
sns.distplot(data['N'],color='hotpink')
plt.xlabel("Nitrogen",fontsize=20,labelpad=5,color='hotpink')
plt.grid()

plt.subplot(2,4,2)
sns.distplot(data['P'],color='yellow')
plt.xlabel("Phosphorus",fontsize=20,labelpad=5,color='yellow')
plt.grid()

plt.subplot(2,4,3)
sns.distplot(data['K'],color='red')
plt.xlabel("Pottasium",fontsize=20,labelpad=5,color='red')
plt.grid()

plt.subplot(2,4,4)
sns.distplot(data['temperature'],color='blue')
plt.xlabel("Temperature",fontsize=20,labelpad=5,color='blue')
plt.grid()

plt.subplot(2,4,5)
sns.distplot(data['ph'],color='grey')
plt.xlabel("PH Value",fontsize=20,labelpad=5,color='grey')
plt.grid()

plt.subplot(2,4,6)
sns.distplot(data['humidity'],color='green')
plt.xlabel("Humidity",fontsize=20,labelpad=5,color='green')
plt.grid()

plt.subplot(2,4,7)
sns.distplot(data['rainfall'],color='violet')
plt.xlabel("Rainfall",fontsize=20,labelpad=5,color='violet')
plt.grid()

plt.show()
```
